```python
# imports
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import ogcore.output_plots as op
import ogcore.output_tables as ot
from ogcore.utils import safe_read_pickle

# set plot style
plt.style.use("ogcore.OGcorePlots")
# set current directory
CUR_DIR = os.getcwd()
```

```python
# Set path for simulation to use
sim_path = os.path.join(CUR_DIR, "OG-PHL-UBI-WEALTH")
```

```python
# read in output from health ins simulation
base_params = safe_read_pickle(
    os.path.join(
        sim_path, "OUTPUT_BASELINE", "model_params.pkl"
        )
    )
base_tpi = safe_read_pickle(
    os.path.join(
        sim_path, "OUTPUT_BASELINE", "TPI", "TPI_vars.pkl"
        )
    )
reform_params = safe_read_pickle(
    os.path.join(
        sim_path, "OUTPUT_REFORM", "model_params.pkl"
        )
    )
reform_tpi = safe_read_pickle(
    os.path.join(
        sim_path, "OUTPUT_REFORM", "TPI", "TPI_vars.pkl"
        )
    )
```

# Efficiency

```python
# use a op function to plot pct diff in GDP
fig = op.plot_aggregates(
    base_tpi,
    base_params,
    reform_tpi=reform_tpi,
    reform_params=reform_params,
    var_list=["Y", "L", "K"],
    plot_type="pct_diff",
    num_years_to_plot=30,
    start_year=base_params.start_year +2
    )
fig.show()
```

# Fiscal Impacts

```python
# debt to GDP
fig = op.plot_gdp_ratio(
    base_tpi,
    base_params,
    reform_tpi=reform_tpi,
    reform_params=reform_params,
    var_list=["D"],
    plot_type="diff",
    num_years_to_plot=50,
    start_year=base_params.start_year,
)
fig.show()
```

```python
# tax revenue to GDP
fig = op.plot_gdp_ratio(
    base_tpi,
    base_params,
    reform_tpi=reform_tpi,
    reform_params=reform_params,
    var_list=["total_tax_revenue"],
    plot_type="levels",
    num_years_to_plot=50,
    start_year=base_params.start_year+ 2,
)
fig.show()
```

# Inequality

```python
fig = op.inequality_plot(
    base_tpi,
    base_params,
    reform_tpi=reform_tpi,
    reform_params=reform_params,
    var="bmat_splus1",
    ineq_measure="gini",
    pctiles=None,
    plot_type="levels",
    num_years_to_plot=50,
    start_year=base_params.start_year +2,
)
fig.show()
```

```python
fig = op.inequality_plot(
    base_tpi,
    base_params,
    reform_tpi=reform_tpi,
    reform_params=reform_params,
    var="bmat_splus1",
    ineq_measure="top_share",
    pctiles=0.01,
    plot_type="levels",
    num_years_to_plot=50,
    start_year=base_params.start_year +2,
)
fig.show()
```

```python
# Plot income by percentile
fig = op.ability_bar(
    base_tpi,
    base_params,
    reform_tpi,
    reform_params,
    var="c_path",
    num_years=5,
    start_year=base_params.start_year + 30,
)
fig.show()
```
